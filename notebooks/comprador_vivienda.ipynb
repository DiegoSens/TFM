```python
#Importing packages.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas_profiling as pdp
plt.style.use('seaborn')
import seaborn as sns
```

## Total 2020

```python
#Loading data.
xls = pd.ExcelFile('../data/comprador_vivienda_ccaa_final.xlsx')
total_2019 = pd.read_excel(xls, 'Total2019')
total_2020 = pd.read_excel(xls, 'Total2020')
primer_trimestre_2021 = pd.read_excel(xls, '1t 2021')
total_2020
```

```python
total_2020 = total_2020.set_index(total_2020['Territorio'])
total_2020.drop(['Territorio'], axis='columns', inplace=True)
total_2020
```

```python
ax = sns.heatmap(total_2020, linewidths=.5)
```

```python
fig = plt.figure(figsize=(20,10))
ax = sns.heatmap(total_2020, annot=True, fmt="d")
```

```python
fig = plt.figure(figsize=(27,17))

x = plt.xticks(fontsize=25, weight="bold")
y = plt.yticks(fontsize=25, weight="bold")

ax = sns.heatmap(total_2020, annot=True, fmt="d", cmap="YlGnBu", annot_kws={'size':24}, cbar=False)
```
