```python
#Importing packages.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas_profiling as pdp
plt.style.use('seaborn')
```

```python
#Loading data.
xls = pd.ExcelFile('../data/data_compraventas.xlsx')
compraventas = pd.read_excel(xls, 'data')
compraventas
```

```python
compraventas.info()
```

```python
f0 = compraventas[' '] == "Total Nacional"
españa_compraventas = compraventas[f0]

del(españa_compraventas[" "])
españa_compraventas = españa_compraventas.transpose()
```

```python
españa_compraventas
```

```python
fig = plt.figure(figsize=(40,20))

ax1 = plt.plot(españa_compraventas, marker='', markerfacecolor='blue', markersize=1, color='firebrick', 
          linewidth=10)

ax1 = plt.xlabel("Mes", fontsize=70, weight="bold")
x = plt.xticks(fontsize=40, rotation=50, weight="bold")
ax1 = plt.ylabel("Número de hipotecas", fontsize=70, weight="bold")
y = plt.yticks(fontsize=60, weight="bold")

ax1 = plt.legend(['Compraventas de viviendas en España'],prop={"size":70, 'weight':'bold'}, loc="lower right")

plt.vlines("Marzo_2020", 20000, 50000, linestyles='dashed', colors='black', linewidth=6)

plt.grid(True)
plt.show()
```

```python
f0 = compraventas[' '] == "13 Madrid, Comunidad de"
madrid = compraventas[f0]

del(madrid[" "])
madrid = madrid.transpose()
```

```python
madrid
```

```python
f0 = compraventas[' '] == "09 Cataluña"
cataluña = compraventas[f0]

del(cataluña[" "])
cataluña = cataluña.transpose()
```

```python
cataluña
```

```python
fig = plt.figure(figsize=(40,20))

plt.plot(madrid, marker='', markerfacecolor='blue', markersize=1, color='dodgerblue', 
          linewidth=10, label = 'Número de compraventas de viviendas - Comunidad de Madrid')

plt.plot(cataluña, marker='', markerfacecolor='blue', markersize=1, color='darkorange', 
          linewidth=10, label = 'Número de compraventas de viviendas - Cataluña')

ax1 = plt.xlabel("Mes", fontsize=50)
x = plt.xticks(fontsize=30, rotation=50)
ax1 = plt.ylabel("Número de compraventas", fontsize=50)
y = plt.yticks(fontsize=30)

ax1 = plt.legend(prop={"size":40}, loc="lower right")

plt.vlines("Marzo_2020", 3500, 8000, linestyles='dashed', colors='black', linewidth=6)

plt.grid(True)
plt.show()
```
