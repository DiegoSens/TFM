```python
#Importing packages.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas_profiling as pdp
plt.style.use('seaborn')
```

```python
#Loading data.
xls = pd.ExcelFile('../data/data_intereses_hipotecas.xlsx')
tipo = pd.read_excel(xls, 'data')
tipo
```

```python
f0 = tipo['Tipo'] == "Total"
total = tipo[f0]

del(total["Tipo"])
total = total.transpose()

total
```

```python
f0 = tipo['Tipo'] == "Fijo"
fijo = tipo[f0]

del(fijo["Tipo"])
fijo = fijo.transpose()

fijo
```

```python
f0 = tipo['Tipo'] == "Variable"
variable = tipo[f0]

del(variable["Tipo"])
variable = variable.transpose()

variable
```

```python
fig = plt.figure(figsize=(40,20))

plt.plot(total, marker='', markerfacecolor='blue', markersize=1, color='darkgreen', 
          linewidth=8, label="Tipo medio total")
plt.plot(fijo, marker='', markerfacecolor='blue', markersize=1, color='c', 
          linewidth=8, label = "Tipo medio de las hipotecas fijas")
plt.plot(variable, marker='', markerfacecolor='blue', markersize=1, color='blue', 
          linewidth=8, label = "Tipo medio de las hipotecas variables")

ax1 = plt.xlabel("Mes", fontsize=50, weight="bold")
x = plt.xticks(fontsize=50, rotation=50, weight="bold")
ax1 = plt.ylabel("Tipo de interés", fontsize=50, weight="bold")
y = plt.yticks(fontsize=50, weight="bold")

ax1 = plt.legend(prop={"size":55, 'weight':'bold'}, loc="lower left")

plt.xticks([0,12,24,36,48,60,72,84,96,108,120,132,144])

plt.vlines("2020M03", 2, 7, linestyles='dashed', colors='black', linewidth=6)

plt.grid(True)
plt.show()
```
