```python
#Importing packages.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas_profiling as pdp
import seaborn as sns
plt.style.use('seaborn')
```

```python
#Loading data.
xls = pd.ExcelFile('../data/data_cemento.xlsx')
cemento = pd.read_excel(xls, 'NACIONAL')
cemento
```

```python
cemento.info()
```

```python
cemento = cemento.set_index('PERIODO')
cemento
```

```python
plt.figure(figsize=(70,30))

plt.plot(cemento.iloc[:,[1]], marker='', markerfacecolor='red', markersize=12, color='red', 
         linewidth=8, label = 'Producción de Cemento')

plt.plot(cemento.iloc[:,[4]], marker='', markerfacecolor='red', markersize=12, color='purple', 
         linewidth=8, label = 'Total Ventas Domésticas de Cemento')

ax1 = plt.xlabel("Mes", fontsize=70, weight="bold")
x = plt.xticks(fontsize=65, rotation=50, weight="bold")
ax1 = plt.ylabel("Millones de toneladas", fontsize=70, weight="bold")
y = plt.yticks(fontsize=65, weight="bold")

plt.vlines("Marzo_2020", 500000, 1750000, linestyles='dashed', colors='black', linewidth=6)

ax1 = plt.legend(prop={"size":70, 'weight':'bold'}, loc="lower right")

plt.savefig('Producción de Cemento');
```

```python
plt.figure(figsize=(70,30))
plt.plot(cemento.iloc[:,[8]], marker='', markerfacecolor='red', markersize=12, color='green', 
         linewidth=8, label = 'Total Exportaciones de Cemento y Clinker')

plt.plot(cemento.iloc[:,[11]], marker='', markerfacecolor='red', markersize=12, color='blue', 
         linewidth=8, label = 'Total Importaciones de Cemento y Clinker')

ax1 = plt.xlabel("Mes", fontsize=70, weight="bold")
x = plt.xticks(fontsize=65, rotation=50, weight="bold")
ax1 = plt.ylabel("Toneladas", fontsize=70, weight="bold")
y = plt.yticks(fontsize=65, weight="bold")

plt.vlines("Marzo_2020", 40000, 700000, linestyles='dashed', colors='black', linewidth=6)

ax1 = plt.legend(prop={"size":65, 'weight':'bold'}, loc="upper left")

plt.savefig('Importaciones y Exportaciones de Cemento');
```
