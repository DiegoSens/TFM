```python
#Importing packages.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas_profiling as pdp
plt.style.use('seaborn')
import seaborn as sns
```

```python
#Loading data.
xls = pd.ExcelFile('../data/nacional_covid.xlsx')
casos_españa = pd.read_excel(xls, 'data')
casos_españa
```

```python
casos_españa.info()
```

```python
casos_españa = casos_españa.set_index(casos_españa['fecha'])
casos_españa.drop(['fecha'], axis='columns', inplace=True)
casos_españa
```

```python
plt.plot(casos_españa)
```

```python
f, ax = plt.subplots(figsize=(15, 5))
sns.lineplot(x="fecha", y="casos_total", data=casos_españa)

plt.title('Hospitalizados en planta y en UCI - Castilla y León')  #Título del histograma

plt.xlabel('Fecha')  #Título del eje X
plt.ylabel('Hospitalizados en planta y UCI')  #Título del eje Y

plt.show()
```

```python
fig = plt.figure(figsize=(60,30))

ax1 = plt.plot(casos_españa, marker='', markerfacecolor='blue', markersize=1, color='darkcyan', 
          linewidth=12)

ax1 = plt.xlabel("Fecha", fontsize=70, weight="bold")
x = plt.xticks(fontsize=70, rotation=50, weight="bold")
ax1 = plt.ylabel("Número de contagios", fontsize=70, weight="bold")
y = plt.yticks(fontsize=70, weight="bold")

ax1 = plt.legend(['Número de contagios diarios en España'],prop={"size":80, 'weight':'bold'}, loc="upper right")

plt.grid(True)
plt.show()
```
