```python
#Importing packages.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas_profiling as pdp
plt.style.use('seaborn')
```

```python
#Loading data.
xls = pd.ExcelFile('../data/tipo_vivienda.xlsx')
tipo = pd.read_excel(xls, 'data')
tipo
```

```python
tipo.info()
```

```python
f0 = tipo['Tipo'] == "Vivienda libre Madrid"
libre = tipo[f0]
```

```python
del(libre["Tipo"])
libre = libre.transpose()
```

```python
libre
```

```python
f0 = tipo['Tipo'] == "Vivienda Segunda Mano Madrid"
segunda = tipo[f0]
```

```python
del(segunda["Tipo"])
segunda = segunda.transpose()
```

```python
segunda
```

```python
f0 = tipo['Tipo'] == "Vivienda nueva Madrid"
nueva = tipo[f0]
```

```python
del(nueva["Tipo"])
nueva = nueva.transpose()
```

```python
nueva
```

```python
f0 = tipo['Tipo'] == "Vivienda Protegida Madrid"
protegida = tipo[f0]
```

```python
del(protegida["Tipo"])
protegida = protegida.transpose()
```

```python
protegida
```

# Gr√°ficos

```python
fig = plt.figure(figsize=(40,20))

plt.plot(libre, marker='', markerfacecolor='blue', markersize=1, color='darkgreen', 
          linewidth=10, label="N√∫mero de viviendas libres - Madrid")
plt.plot(segunda, marker='', markerfacecolor='blue', markersize=1, color='c', 
          linewidth=10, label = "N√∫mero de viviendas de segunda mano - Madrid")

ax1 = plt.xlabel("Trimestre", fontsize=50, weight="bold")
x = plt.xticks(fontsize=50, rotation=50, weight="bold")
ax1 = plt.ylabel("N√∫mero de viviendas", fontsize=50, weight="bold")
y = plt.yticks(fontsize=50, weight="bold")

ax1 = plt.legend(prop={"size":55, 'weight':'bold'}, loc="lower left")

plt.vlines("2020T1", 4000, 12000, linestyles='dashed', colors='black', linewidth=6)

plt.grid(True)
plt.show()
```

```python
fig = plt.figure(figsize=(40,20))

plt.plot(nueva, marker='', markerfacecolor='blue', markersize=1, color='salmon', 
          linewidth=10, label="N√∫mero de viviendas nuevas - Madrid")
plt.plot(protegida, marker='', markerfacecolor='blue', markersize=1, color='peru', 
          linewidth=10, label = "N√∫mero de viviendas protegidas - Madrid")

ax1 = plt.xlabel("Trimestre", fontsize=50, weight="bold")
x = plt.xticks(fontsize=50, rotation=50, weight="bold")
ax1 = plt.ylabel("N√∫mero de viviendas", fontsize=50, weight="bold")
y = plt.yticks(fontsize=50, weight="bold")

ax1 = plt.legend(prop={"size":55, 'weight':'bold'}, loc="upper left")

plt.vlines("2020T1", 150, 1850, linestyles='dashed', colors='black', linewidth=6)

plt.grid(True)
plt.show()
```
