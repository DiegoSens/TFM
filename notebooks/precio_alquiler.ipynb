```python
#Importing packages.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas_profiling as pdp
plt.style.use('seaborn')
```

```python
#Loading data.
xls = pd.ExcelFile('../data/precio_alquiler.xlsx')
precio_alquiler = pd.read_excel(xls, 'data')
precio_alquiler
```

```python
precio_alquiler.info()
```

```python
precio_alquiler.describe()
```

## Madrid

```python
f0 = precio_alquiler['Zona'] == "     Madrid"
madrid = precio_alquiler[f0]
madrid
```

```python
del(madrid["Zona"])
madrid = madrid.transpose()
madrid
```

```python
madrid.describe()
```

```python
plt.figure(figsize=(80,40))

plt.plot(madrid, marker='', markerfacecolor='blue', markersize=1, color='dodgerblue', 
          linewidth=16, label = 'Precio medio del alquiler de la vivienda en la Comunidad de Madrid')

ax1 = plt.xlabel("Mes", fontsize=80, weight="bold")
x = plt.xticks(fontsize=70, rotation=50, weight="bold")
ax1 = plt.ylabel("Euros/m2", fontsize=70, weight="bold")
y = plt.yticks(fontsize=70, weight="bold")

ax1 = plt.legend(prop={"size":79, 'weight':"bold"}, loc="upper left")

plt.vlines("marzo_2020", 14.5, 17, linestyles='dashed', colors='black', linewidth=10)

plt.xticks([0,3,6,9,12,15,18,21,24,27,30,33,36,39])

plt.grid(True)
plt.show()
```
