# Librerías

```python
#Importing packages.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# import lux
# # Hide warnings
# import warnings
# warnings.filterwarnings('ignore')
import pandas_profiling as pdp
plt.style.use('seaborn')
```

# Datos

```python
#Loading data.
xls = pd.ExcelFile('../data/migraciones.xls')
migraciones = pd.read_excel(xls, 'data')
migraciones
```

```python
migraciones.set_index(migraciones.año, inplace=True, drop=True)
del(migraciones["año"])
migraciones
```

## Migraciones entre Madrid y otras CCAAs

```python
fig = plt.figure(figsize=(40,20))

plt.plot(migraciones.migraciones_ccaa_a_madrid, marker='', markerfacecolor='blue', markersize=1, color='dodgerblue', 
          linewidth=10, label = 'Migraciones de otras CCAA a la C.Madrid')

plt.plot(migraciones.migraciones_madrid_a_ccaa, marker='', markerfacecolor='blue', markersize=1, color='darkorange', 
          linewidth=10, label = 'Migraciones de la C.Madrid a otras CCAA')

ax1 = plt.xlabel("Año", fontsize=50, weight="bold")
x = plt.xticks(fontsize=50, rotation=50, weight="bold")
ax1 = plt.ylabel("Número de migraciones", fontsize=50, weight="bold")
y = plt.yticks(fontsize=50, weight="bold")

ax1 = plt.legend(prop={"size":50, 'weight':'bold'}, loc="upper left")

plt.vlines(2020, 45000, 85000, linestyles='dashed', colors='black', linewidth=6)

plt.grid(True)
plt.show()
```

## Migraciones entre Madrid y otras países extranjeros

```python
fig = plt.figure(figsize=(40,20))

plt.plot(migraciones.migraciones_ext_a_madrid, marker='', markerfacecolor='blue', markersize=1, color='indigo', 
          linewidth=10, label = 'Migraciones desde el extranjero a la C.Madrid')

plt.plot(migraciones.migraciones_madrid_a_ext, marker='', markerfacecolor='blue', markersize=1, color='lightpink', 
          linewidth=10, label = 'Migraciones de la C.Madrid al extranjero')


ax1 = plt.xlabel("Año", fontsize=50, weight="bold")
x = plt.xticks(fontsize=50, rotation=50, weight="bold")
ax1 = plt.ylabel("Número de migraciones", fontsize=50, weight="bold")
y = plt.yticks(fontsize=50, weight="bold")

ax1 = plt.legend(prop={"size":50, 'weight':'bold'}, loc="upper left")

plt.vlines(2020, 4500, 25000, linestyles='dashed', colors='black', linewidth=6)

plt.grid(True)
plt.show()
```
